```python
import subprocess
import sys

def run_on_pico(code: str, port: str = "COM3"):
    cmd = [
        sys.executable, "-m", "mpremote",
        "connect", port,
        "exec", code
    ]

    proc = subprocess.run(cmd, capture_output=True, text=True, encoding="utf-8")

    print("STDOUT:")
    print(proc.stdout)

    print("STDERR:")
    print(proc.stderr)
```

```python
code = """
from machine import Pin
from onewire import OneWire
from ds18x20 import DS18X20
import time

ow = OneWire(Pin(22))
ds = DS18X20(ow)

devices = ds.scan()

print("Devices:", devices)

ds.convert_temp()
time.sleep_ms(750)

for dev in devices:
    temp = ds.read_temp(dev)
    print("Temperatur:", temp, "°C")
"""
```

```python
run_on_pico(code)
```
